```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 선그래프
data1 = [10, 14, 19, 20, 25]  # y축 데이터, x죽 데이터 설정하지 않으면 자동으로 0부터 1씩 증가하는 값으로 할당
plt.plot(data1)  # 선그래프 그리기
```

```python
x = np.arange(-4.5, 5, 0.5)  # x축 데이터
y = 2 * x**2  # y축 데이터
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
# 그래프 여러개 그리기
y2 = 5 * x + 30
y3 = 4 * x **2 + 10
print(x)
print(y)
print(y2)
print(y3)
```

```python
plt.plot(x, y, x, y2, x, y3)  # 여러개의 x데이터와 y데이터의 쌍을 입력한다.
plt.show()
```

```python
# 새로운 창에 그래프 생성
plt.plot(x, y)  # 처음엔 그래프창이 자동으로 생성됨
plt.figure()  # 새롭게 생성된 그래프 창에 그래프를 그림
plt.plot(x, y2)
plt.show()
```

```python
# 그래프 창의 번호를 지정해서 해당 창에 그래프 그리기
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20 * np.cos(x)**2  # cos()는 numpy의 np.cos()을 사용

plt.figure(1)  # 1번 그래프 창을 생성
plt.plot(x, y1)  # 지정된 그래프 창에 그래프를 그림

plt.figure(2)  # 2번 그래프 창을 생성
plt.plot(x, y2)

plt.figure(1)  # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2)

plt.figure(2)  # 이미 생성된 1번 그래프 창을 지정함
plt.clf()  # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1)  # 지정된 2번 그래프 창에 그래프를 그림

plt.show()
```

```python
# 하나의 그래프 창을 하위 그래프 영역으로 나누기
x = np.arange(0, 10, 0.1)
y1 = 0.3 * (x - 5)**2 + 1
y2 = -1.5 * x + 3
y3 = np.sin(x)**2  # sin()는 numpy의 np.sin()을 사용
y4 = 10 * np.exp(-x) + 1  # exp()는 numpy의 np.exp()을 사용

# 2x2 행렬로 이루어진 하위 그래프에서 위치를 지정
plt.subplot(2, 2, 1)  # 첫번째
plt.plot(x, y1)

plt.subplot(2, 2, 2)  # 두번째
plt.plot(x, y2)

plt.subplot(2, 2, 3)  # 세번째
plt.plot(x, y3)

plt.subplot(2, 2, 4)  # 네번째
plt.plot(x, y4)

plt.show()
```

```python
# 좌표 범위를 지정해 관심 영역만 그래프로 그리기
x = np.linspace(-4, 4, 100)  # [-4, 4] 범위에서 100개의 값 생성
y1 = x**3
y2 = 10 * x ** 2 - 2

plt.plot(x, y1, x, y2)  # 두개의 그래프 한 창에 생성
plt.show()
```

```python
plt.plot(x, y1, x, y2)
plt.xlim(-1, 1)  # 범위를 지정해 영역 확대
plt.ylim(-3, 3)
plt.show()
```
