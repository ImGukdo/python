```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://google.com')  # html 소스를 위한 응답 객체
print(res.status_code)  # 200 이면 정상
```

```python
res.raise_for_status()  # 200이 아닌 결과에 대해서 raise를 발생 시켜줌 (401, 403 등)
soup = BeautifulSoup(res.text, 'lxml')  # BeautifulSoup을 이용해 html 소스를 파싱
soup
```

```python
soup.find('a')  # a 태그가 있는 첫 번째 요소를 찾아서 반환
```

```python
soup.find('a').get_text()  # 요소에서 텍스트 문자열만 반환
```

```python
soup.find('a').string  # 요소에서 텍스트 문자열만 반환
```

```python
links = soup.find_all('a')  # a 태그로 시작하는 모든 요소를 다 반환
links
```

```python
for i in links :
    print(i.get_text() + ' : ' + i["href"])  # a 태그안에 있는 정보 출력
```

```python
html = '<html><body><p class=a>aaa</p><p class=b>bbb</p><p class=a>ccc</p><p class=a id=xxx>ddd</p></body></html>'
root = BeautifulSoup(html, 'html.parser') 
p1 = root.find_all('p')
print('p1:', p1)
p2 = root.find_all('p', {'class':'a'})
print('p2:', p2)
p3 = root.find_all('p', {'class':'b'})
print('p3:', p3)
p4 = root.find('p', {'class':'a'})
print('p4:', p4)
p5 = root.select('p')
print('p5:', p5)
p6 = root.select('p.a')#태그.클래스명 //태그#id명
print('p6:', p6)
p7 = root.select('p.b')
print('p7:', p7)
p8 = root.select('p#xxx')
print('p8:', p8)
```

```python
html = requests.get('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp').text  #웹 요청
#html=html.decode('utf-8')
#지역별 일간 날씨를 출력
root = BeautifulSoup(html, 'lxml') 
root
```

```python
loc = root.select('location > data')  # location 태그 안에 data 태그들을 가져온다.
print(loc[0].tmef.string, loc[0].wf.string)
```
